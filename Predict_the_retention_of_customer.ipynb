```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
data={"Age":[23,45,43,54,26,43,19],"MonthlyRecharge":[100,110,120,50,90,55,45],"churn":[1,0,1,0,1,0,1]}
df=pd.DataFrame(data)
df
```

```python
x = df[["Age","MonthlyRecharge"]]
y = df["churn"]
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
model=SVC(kernel="linear",C=0.1)# c = Default Regularisation
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
report=classification_report(y_test,y_pred)
print(report)
```

```python
user_age=float(input("Enter your age: "))
user_monthly_recharge=float(input("Enter your recharge amount: "))
user_input=np.array([[user_age,user_monthly_recharge]])
prediction=model.predict(user_input)
print(f"prediction:{prediction[0]}")
if prediction[0]==1:
  print("user is likely to churn")
else:
  print("user is not likely to churn")

```
