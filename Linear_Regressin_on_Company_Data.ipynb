```python
# On the basis of Years of startup open, Analyse their profit.

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("/content/drive/MyDrive/Company_Profit.csv")
print(df)
```

```python
graph=sns.pairplot(df)
print(graph)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,1:].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
r2u=r2_score(y_test,y_pred)
print(mse)
print(r2u)
```

```python
accuracy=model.score(y_test,y_pred)
print(accuracy)
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,color='red')
plt.show()
```

```python
user_startup_year = float(input("Enter your startup time: "))
user_profit = model.predict([[user_startup_year]])
print(f"User profit: {user_profit[0][0]:.2f}")
```
